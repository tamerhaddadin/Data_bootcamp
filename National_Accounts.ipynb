```python
import numpy as np
import pandas as pd
```

```python
#######------------------ Computing capital income share and the capital/output for 5 countries from Penn World Table ------------#######
 
## Step1: Reading the excel file ##

na = pd.read_excel('pwt100.xlsx',sheet_name='Data')


```

```python
## S2: Keeping the varaibles I will use from the table ##
na = na[['country','year','countrycode','rgdpna','rnna','irr']]
```

```python
## S3: Creating the variables of interest ##

# Creating the capital (K) to output (Y) ratio column #
na['K/Y'] = na['rnna']/na['rgdpna']

# Creating a column for capital income share (sK) #
na['sK']= na['K/Y']*na['irr']
```

```python

## S4: Choosing the countries and the two periods of time ##

# Keeping Japan, Mexico, Morocco, Mauritius and Kenya #
na= na[((na['countrycode'] == 'JPN') | (na['countrycode'] == 'MEX') | (na['countrycode'] == 'MUS') | (na['countrycode'] == 'KEN')|(na['countrycode'] == 'MAR'))]
na.head(6)

# Keeping 1950 and 2015 # 

na = na [(na['year'] == 1950) | (na['year'] == 2015)]


na

```
